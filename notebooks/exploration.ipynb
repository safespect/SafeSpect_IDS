```python
import pandas as pd
from pathlib import Path

RAW_DATA_DIR = Path("../data/raw")

# Load just ONE file first
csv_files = list(RAW_DATA_DIR.glob("*.csv"))
print("Total CSV files found:", len(csv_files))

sample_file = csv_files[0]
print("Inspecting:", sample_file.name)

df = pd.read_csv(sample_file)
```

```python
print("Shape (rows, columns):", df.shape)
print("\nColumn names:\n", df.columns.tolist())
print("\nData types:\n", df.dtypes)
```

```python
label_candidates = [col for col in df.columns if "label" in col.lower()]
label_candidates
```

```python
import numpy as np

print("Missing values per column:")
print(df.isna().sum().sort_values(ascending=False).head(10))

print("\nInfinity values count:")
print(np.isinf(df.select_dtypes(include=[float, int])).sum().sum())
```

```python
from pathlib import Path
import pandas as pd

RAW_DATA_DIR = Path("../data/raw")

all_columns = None
all_labels = set()

for file in RAW_DATA_DIR.glob("*.csv"):
    df_temp = pd.read_csv(file)
    
    # Check column consistency
    if all_columns is None:
        all_columns = set(df_temp.columns)
    else:
        if set(df_temp.columns) != all_columns:
            print("Column mismatch in:", file.name)
    
    # Collect labels
    all_labels.update(df_temp[" Label"].unique())

print("All unique labels across dataset:")
print(all_labels)
```
